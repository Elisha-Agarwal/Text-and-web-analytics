## Q1 Find frequency of each word in file

```python
import nltk
from nltk.corpus import gutenberg
from nltk.probability import FreqDist
```

```python
nltk.download('gutenberg')
wt_words = gutenberg.words('shakespeare-hamlet.txt')
data_analysis = nltk.FreqDist(wt_words)
```

```python
# Let's take the specific words only if their frequency is greater than 2.
filter_words = dict([(m, n) for m, n in data_analysis.items() if len(m) > 2])
 
for key in sorted(filter_words):
    print("%s: %s" % (key, filter_words[key]))
 
data_analysis = nltk.FreqDist(filter_words)
 
data_analysis.plot(25, cumulative=False)
```

## Q2 Find the most occuring words

```python
from collections import Counter


#word_list = word_set.split()


word_count = Counter(filter_words)

# Use most_common() method from Counter subclass
print(word_count.most_common(5))
```
